#Gráfico de Linhas

Gráficos de linhas são ideais para mostrar alterações e tendências de dados em intervalos de tempo iguais.

<center><img src="https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_03/grafico_linha.png"width="400px"></center>


Quando se quer observar o comportamento dos dados ao longo do tempo, usa-se o eixo x (horizontal) para representar a ordem cronológica e o eixo y para a variável dependente.

#Plotar Gráfico de Linhas
Como exemplo de gráfico de linhas vou usar um dataset contendo o número de casos (mensal) de dengue para a cidade de Campinas-SP.

Além do número de casos confirmados de dengue, o dataset mostra ainda dados relacionados à chuva, temperaturas média, mínima e máxima do município. O período contemplado pelo conjunto de dados é de 1998 a 2015.

```
# importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

#importar os dados para dataframe
data = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/dengue-dataset.csv"
df = pd.read_csv(data)
#coluna data é um objeto do tipo datetime
df.data = pd.to_datetime(df.data)

#qndo temos eventos ocorrendo em datas discriminadas, colocamos no indice
df.set_index('data', inplace=True)

# vendo as 5 primeiras entradas
df.head()
```

```
# plotar o numero de casos confirmados
fig, ax = plt.subplots()
ax.plot('casos-confirmados', data=df);
ax.set_title("Casos de Dengue em Campinas")
ax.set_xlabel("Data")
ax.set_ylabel("Casos confirmados")
#vai inclinar a data
fig.autofmt_xdate()

plt.show()
```

```
# plotar o mumero de casos temperatura média e temperatura máxima
fig, ax= plt.subplots()
ax.plot('temperatura-media', data=df)
ax.plot('temperatura-maxima', data=df)
ax.set_title("Casos confirmados de Dengue em Campinas")
ax.set_xlabel("Data")
ax.set_ylabel("Casos confirmados")
fig.autofmt_xdate()

plt.show()
```

#Selecionando intervalos


```
# plotar o histograma da temperatura média '2011-01-01':'2013-12-31'
fig, ax = plt.subplots()

#Selecionar intervalo desejado
df_sel = df.loc['2011-01-01':'2013-12-31']
ax.set_title("Casos Confirmados de Dengue em Campinas")
ax.set_xlabel("Data")
ax.set_ylabel("Casos Confirmados")
ax.plot('casos-confirmados', data=df_sel)
#arrumando as datas
fig.autofmt_xdate()
```

#Plotar mútiplos gráficos separdamente

```
# plotar um histograma da temperatura média
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 4))

#plotar temperatura média
ax[0].plot('temperatura-media', data=df)
ax[0].set_title("Temperatura Média")

#plotar casos confirmados
ax[1].plot('casos-confirmados', data=df)
ax[1].set_title("Casos Confirmados")

#para uma visualização melhor usamos, ele trás um layot na posição maiss desejada 
plt.tight_layout()
```
